# Bulk Query Export

Exports raw Tideway report queries through the public reports API.


## Setup

Raw query export writes multiple files, so `OUTPUT_DIR` is required.

```python
from pathlib import Path

import pandas as pd
import tideway
from tideway import notebooks as tw_nb

APPLIANCE_NAME = None
APPLIANCE_INDEX = 0
WRITE_OUTPUT = False
OUTPUT_BASE_DIR = None

tw = tw_nb.appliance_from_config(appliance_name=APPLIANCE_NAME, appliance_index=APPLIANCE_INDEX)
output_dir = tw_nb.output_dir_for(tw.target, OUTPUT_BASE_DIR)
print("Target      :", tw.target)
print("Output dir  :", output_dir)

OUTPUT_DIR = output_dir
```

## Export Queries

```python
REPORT_NAMES = None  # None/default exports the default query set

result = tw.reports().run(names=REPORT_NAMES, queries=True, output_dir=str(OUTPUT_DIR))
print("Files:")
for path in result.files:
    print(" -", path)
```
