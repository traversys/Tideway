# Lifecycle Reports

Runs the OS, software, and database lifecycle reports through Tideway.


## Setup

Set `WRITE_OUTPUT = True` to write canonical report files.

```python
from pathlib import Path

import pandas as pd
import tideway
from tideway import notebooks as tw_nb

APPLIANCE_NAME = None
APPLIANCE_INDEX = 0
WRITE_OUTPUT = False
OUTPUT_BASE_DIR = None

tw = tw_nb.appliance_from_config(appliance_name=APPLIANCE_NAME, appliance_index=APPLIANCE_INDEX)
output_dir = tw_nb.output_dir_for(tw.target, OUTPUT_BASE_DIR)
print("Target      :", tw.target)
print("Output dir  :", output_dir)
```

## Run Reports

```python
REPORT_NAMES = ['os_lifecycle', 'software_lifecycle', 'db_lifecycle']
REPORT_OPTIONS = {}

reports = tw.reports()

if WRITE_OUTPUT:
    result = reports.run(names=REPORT_NAMES, output_dir=str(output_dir), **REPORT_OPTIONS)
else:
    result = reports.run(names=REPORT_NAMES, **REPORT_OPTIONS)

for item in result.results:
    print("##", item.name)
    df = tw_nb.report_to_dataframe(item)
    display(df.head(25))
    print("Rows:", len(df))

if getattr(result, "files", None):
    print("Files:")
    for path in result.files:
        print(" -", path)
```
