# Tideway Taxonomy API Examples

Helper methods paired with direct calls for taxonomy endpoints (sections, locales, node kinds, relationship kinds, field lists). Reference: https://docs.bmc.com/xwiki/bin/view/IT-Operations-Management/Discovery/BMC-Helix-Discovery/DAAS/Integrating/Using-the-REST-APIs/Endpoints-in-the-REST-API/.

## Setup
- Install tideway (e.g. `pip install -e .` from repo root).
- Set `TARGET` and `TOKEN` below. Do **not** commit secrets.
- Adjust filters (section/locale/kind) as needed.

```python

import tideway

# Configuration
TARGET = 'appliance-hostname-or-ip'  # e.g. 'discovery.example.com'
TOKEN = 'your-api-token'            # keep secrets out of source control
API_VERSION = '1.16'                # latest supported API
SSL_VERIFY = False                  # set to True when using valid certs

tw = tideway.appliance(TARGET, TOKEN, api_version=API_VERSION, ssl_verify=SSL_VERIFY)
taxonomy = tw.taxonomy()

def show_response(resp, label):
    if resp.ok:
        try:
            return resp.json()
        except Exception:
            return resp.text
    try:
        body = resp.json()
    except Exception:
        body = resp.text
    return {'endpoint': label, 'status': resp.status_code, 'body': body}

tw.api_about.json()  # quick sanity check
```

## Sections and locales
List taxonomy sections/locales via helper vs direct GET.

```python

sections_helper = taxonomy.get_taxonomy_sections
sections_direct = taxonomy.get('/taxonomy/sections')

locales_helper = taxonomy.get_taxonomy_locales
locales_direct = taxonomy.get('/taxonomy/locales')

meta_calls = {
    '/taxonomy/sections (helper)': show_response(sections_helper, '/taxonomy/sections'),
    '/taxonomy/sections (direct)': show_response(sections_direct, '/taxonomy/sections'),
    '/taxonomy/locales (helper)': show_response(locales_helper, '/taxonomy/locales'),
    '/taxonomy/locales (direct)': show_response(locales_direct, '/taxonomy/locales'),
}
meta_calls
```

## Node kinds
Fetch node kinds (optionally filtered) and specific kind or field lists.

```python

node_format = None   # e.g. 'object'
node_section = None  # e.g. 'foundation'
node_locale = None   # e.g. 'en_US'
node_kind = None     # e.g. 'Host'
fieldlists = False
fieldlist_name = None  # e.g. 'default'

nodekinds_helper = taxonomy.get_taxonomy_nodekind(format=node_format, section=node_section, locale=node_locale)
nodekinds_direct = taxonomy.get('/taxonomy/nodekinds')

if node_kind:
    kind_helper = taxonomy.get_taxonomy_nodekind(kind=node_kind, locale=node_locale, fieldlists=fieldlists)
    kind_direct = taxonomy.get(f"/taxonomy/nodekinds/{node_kind}" + ('/fieldlists' if fieldlists else ''))
else:
    kind_helper = kind_direct = 'Set node_kind to fetch a specific node kind or fieldlists.'

if node_kind and fieldlist_name:
    fieldlist_helper = taxonomy.get_taxonomy_nodekind_fieldlist(node_kind, fieldlist_name)
    fieldlist_direct = taxonomy.get(f"/taxonomy/nodekinds/{node_kind}/fieldlists/{fieldlist_name}")
else:
    fieldlist_helper = fieldlist_direct = 'Set node_kind and fieldlist_name to fetch a field list.'

node_calls = {
    '/taxonomy/nodekinds (helper)': show_response(nodekinds_helper, '/taxonomy/nodekinds'),
    '/taxonomy/nodekinds (direct)': show_response(nodekinds_direct, '/taxonomy/nodekinds'),
    'specific_kind_helper': show_response(kind_helper, f"/taxonomy/nodekinds/{node_kind}" if isinstance(kind_helper, object) else 'set node_kind'),
    'specific_kind_direct': show_response(kind_direct, f"/taxonomy/nodekinds/{node_kind}" if isinstance(kind_direct, object) else 'set node_kind'),
    'fieldlist_helper': show_response(fieldlist_helper, f"/taxonomy/nodekinds/{node_kind}/fieldlists/{fieldlist_name}" if isinstance(fieldlist_helper, object) else 'set node_kind/fieldlist_name'),
    'fieldlist_direct': show_response(fieldlist_direct, f"/taxonomy/nodekinds/{node_kind}/fieldlists/{fieldlist_name}" if isinstance(fieldlist_direct, object) else 'set node_kind/fieldlist_name'),
}
node_calls
```

## Relationship kinds
Fetch relationship kinds (optionally filtered) and specific kind details.

```python

rel_format = None  # e.g. 'object'
rel_locale = None  # e.g. 'en_US'
rel_kind = None    # e.g. 'Containment'

relkinds_helper = taxonomy.get_taxonomy_relkind(format=rel_format, locale=rel_locale)
relkinds_direct = taxonomy.get('/taxonomy/relkinds')

if rel_kind:
    relkind_helper = taxonomy.get_taxonomy_relkind(kind=rel_kind, locale=rel_locale)
    relkind_direct = taxonomy.get(f"/taxonomy/relkinds/{rel_kind}")
else:
    relkind_helper = relkind_direct = 'Set rel_kind to fetch a specific relationship kind.'

rel_calls = {
    '/taxonomy/relkinds (helper)': show_response(relkinds_helper, '/taxonomy/relkinds'),
    '/taxonomy/relkinds (direct)': show_response(relkinds_direct, '/taxonomy/relkinds'),
    'relkind_helper': show_response(relkind_helper, f"/taxonomy/relkinds/{rel_kind}" if isinstance(relkind_helper, object) else 'set rel_kind'),
    'relkind_direct': show_response(relkind_direct, f"/taxonomy/relkinds/{rel_kind}" if isinstance(relkind_direct, object) else 'set rel_kind'),
}
rel_calls
```
